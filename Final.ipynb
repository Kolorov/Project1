```python
from colorama import Fore, Back, Style
import time
import speech_recognition as sr
from colorama import init
import pandas as pd
import numpy as np
from subprocess import Popen
import os

import playsound
playsound.playsound('C:/Users/Rohit 2.0/Downloads/voice1.mp3', True)



def SpeechToText():
        r = sr.Recognizer()   #Speech recognition
        with sr.Microphone() as source:
            
            print(Fore.BLUE+"     Listening.....")
            
            audio = r.listen(source)
            message = r.recognize_google(audio)
            
        try:
            print()
            #print("You said: " + r.recognize_google(audio))
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))
        return message
```

```python
def Dbname():
    playsound.playsound('C:/Users/Rohit 2.0/Downloads/database.mp3', True)
    print("What would you like to call your database?")
    name=SpeechToText()
    
    print(Fore.YELLOW+'Did you say ',name,'?')
    ch=SpeechToText()
    if ch=='no' or ch=='NO':
        Dbname()
    elif ch=='yes' or ch=='YES':
        enterData()
        
    else:
        print('Could not understand you Please repeat!\n\n')
        Dbname()
    
    
```

```python
colList=[]
def enterData():
    colList.clear()
    n=6
  
    for i in range(n):
        print(Fore.GREEN+'Enter column',i+1)
        col=SpeechToText()
        colName=col.capitalize()
        colList.append(colName)
    print(Fore.CYAN+'These are your columns right?\n',colList)
    time.sleep(3)
    ch3=SpeechToText()
    if ch3=='no' or ch3=='NO':
        colList.clear()
        print(FORE.CYAN+'\nSpeak again')
        enterData()
         
    elif ch3=='Yes' or ch3=='yes':
        print(Fore.CYAN+'Awesome')
        time.sleep(1)
        createDatabase()
   

    
```
